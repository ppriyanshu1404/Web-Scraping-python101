```python
headings=tables.find_all('th')
```

          text-align:center;">
            Creating Table
</p>

```python
table_row=[]
```

```python
for rows in row_data:
    curr_row=[]
    r_data=rows.find_all('td',{})
    for idx,data in enumerate(r_data):
        curr_row.append(data.text)
    table_row.append(curr_row)
    
    
```

```python
r_data
```

```python
table_row
```

```python
excel_sheet=pd.DataFrame(table_row,columns=['Year','Number of incidents','Death','Injuries','Remarks'])
```

```python
excel_sheet
```

```python
excel_sheet=excel_sheet.drop('Remarks',axis=1)
```

```python
excel_sheet
```

```python
new_sheet=excel_sheet.drop(0,axis=0)
```

```python
new_sheet
```

```python
x=pd.DataFrame(new_sheet)
```
